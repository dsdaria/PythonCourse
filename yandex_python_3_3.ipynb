**3.3.
Списочные выражения. Модель памяти для типов языка Python**

**A. Список квадратов**

```python
[i ** 2 for i in range(a, b + 1)]
```

**B. Таблица умножения 2.0**

```python
[[i * j for j in range(1, n + 1)] for i in range(1, n + 1)]
```

**C. Длины всех слов**

```python
[len(x) for x in sentence.split()]
```

**D. Множество нечетных чисел**

```python
set([element for element in numbers if (element > element % 2 == 1) or (element == 1)])
```

**E. Множество всех полных квадратов**

```python
set([x for x in numbers if (round(x ** 0.5)) ** 2 == x])
```

**F. Буквенная статистика**

```python
{letter: (text.lower()).count(letter) for letter in sorted(text.lower()) if letter.isalpha()}
```

**G. Делители**

```python
{i: [j for j in range(1, i + 1) if i % j == 0] for i in numbers}
```

**H. Аббревиатура**

```python
''.join(x[0].upper() for x in string.split())
```

**I. Преобразование в строку**

```python
' - '.join([str(i) for i in sorted(set(numbers))])
```

**J. RLE наоборот**

```python
''.join([tup[0] * tup[1] for tup in rle])
```
